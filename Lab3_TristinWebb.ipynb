```python
# Author: Tristin Webb
#Created on: 9/11/2025
# Course: CS454 Python Programming II
# Lab3: Data Analysis with NumPy
# Objective: Read a file containing temperature data, store it in a NumPy array, and perform statistical analysis.
# Description: In Task 1 I imported the numpy library and used the genfromtxt method to populate an ndarray.
# Once I complete this step I found the average temperature throughout the year, as well as the max temp and min
# temp. I then used the where function to find which days experience a temperature of 30 degrees and above
# and display the days when this occcured. Lastly I found the standard deviation of this data set.

#Task 1: Analyzing Temperature Data

import numpy as np

path = "temperature_data.csv"

#Load the data from csv into NumPy array

data = np.genfromtxt(path, delimiter=",", encoding = 'utf-8')

#Calculate the Avg, Max, Min, # of days exceeding 30 degrees celcius, and standard deviation

avg_temp = np.average(data[1:, 1])
print('Average: ', avg_temp)

max_temp = np.max(data[1:,1])
print('Maximum: ', max_temp)

min_temp = np.min(data[1:,1])
print('Minimum: ', min_temp)

temps_above_30 = np.where(data[1:, 1] > 30)
print('Days where the temp exceeded 30 degrees (Celsius): ', np.size(temps_above_30))

std_dev = np.std(data[1:, 1])
print('Standard Deviation: ', std_dev)



days_w_max_temp = np.where(data[1:,1] == max_temp)
days_w_max_temp = list(days_w_max_temp[0])
print('List of days with max temp: ', days_w_max_temp)
```

```python
# Author: Tristin Webb
#Created on: 9/11/2025
# Course: CS454 Python Programming II
# Lab3: Data Analysis with NumPy
# Objective: Read demographic data and calculate useful statistics using array-oriented programming.
# Description: 

#Task 3: Exploring Demographic Data
import numpy as np
path = 'population_data.csv'

#Open and read the file
with open(path) as f:
    lines = f.readlines()

#Cleaning the data
updated = [line.strip().split(',') for line in lines]

#Storing the data in a numpy array
census_data = np.array(updated[1:])

#Pulling the population data and converting it to float
population = census_data[:,1].astype(float)

#Reshaping the data and calculating the population per year
population_reshaped = population.reshape(50, 12)
average_pop_by_year = population_reshaped.mean(axis=1)

#Calculating the growth rate by taking final value minus the initial value
growth = average_pop_by_year[-1] - average_pop_by_year[0]

#Calculating the growth rates for each year, and the avg growth rate
growth_rates = ((average_pop_by_year[1:] - average_pop_by_year[:-1]) / average_pop_by_year[:-1]) * 100
avg_annual_growth = growth_rates.mean()

#Two years with the highest growth rate, and the value of growth rate




print('Average population per year:\n', average_pop_by_year)
print('Population growth over 50 years: ', growth)
print('Growth rates:\n', growth_rates)
print('Average yearly growth rate: ', avg_annual_growth) 
```

```python
# Author: Tristin Webb
#Created on: 9/11/2025
# Course: CS454 Python Programming II
# Lab3: Data Analysis with NumPy
# Objective: Analyze a simulated random walk representing stock price fluctuations over time.
# Description: In Task 5 I used the numpy library to use statistical analysis methods such as max, min, mean, cumsum, etc.
# I used the genfromtxt method to load the txt file into a ndarray.

#Task 5: Analysis of Simulated Random Walk for Stock Prices
import numpy as np
path = 'randomwalk_data.csv'

data = np.genfromtxt(path, delimiter = ',', encoding='utf-8')
data[1:,1]

#Daily price changes as a seperate NumPy array
daily_price_changes = np.array(data[1:,1])

#Compute the cumulative sum of daily changes
cumsum = np.cumsum(daily_price_changes)

#Calculate and display the max, min, avg stock price
cumsum = cumsum + 100
max_price = cumsum.max()
min_price = cumsum.min()
avg_price = cumsum.mean()

#print('daily price change:\n', daily_price_changes)
#print('cumulative sum:\n', cumsum)
print('Max Stock Price: $', max_price)
print('Min Stock Price: $', min_price)
print('Avg Stock Price: $', avg_price)
```
